```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
np.random.seed(20)
x_data = np.random.rand(1024)
noise = np.random.normal(scale=0.01, size=x_data.shape)
y_data = x_data*0.1 + noise + 0.2
```

```python
batch_size = 64
```

```python
x_data = pd.DataFrame(x_data)
y_data = pd.DataFrame(y_data)
print(x_data.shape)
print(y_data.shape)
```

#### 仅使用dataset的repeat
发现效果不如epochs效果好，可看

```python
x_placeholder = tf.placeholder(dtype=tf.float32, shape=[None,1])
y_placeholder = tf.placeholder(dtype=tf.float32, shape=[None,1])
dataset = tf.data.Dataset.from_tensor_slices((x_placeholder, y_placeholder))
dataset = dataset.repeat(100)
dataset = dataset.batch(batch_size)
iterator = dataset.make_initializable_iterator()
x_batch, y_batch = iterator.get_next()    # 这个数据要传入到图结构中

w = tf.Variable(np.random.rand(1))
w = tf.cast(w, dtype=tf.float32)
b = tf.Variable(np.random.rand(1))
b = tf.cast(b, dtype=tf.float32)

y_pred = x_batch*w + b

losses = tf.losses.mean_squared_error(y_batch, y_pred)
train = tf.train.GradientDescentOptimizer(0.03).minimize(losses)


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    sess.run(iterator.initializer, feed_dict={x_placeholder: x_data, y_placeholder: y_data})
    i = 1
    while True:
        try:
            i += 1
            print(int(i), end=' ')
            print(sess.run(losses), end='')
            print()
#                 print(sess.run(x_batch))
        except tf.errors.OutOfRangeError:
#             print(i)
            print(sess.run([w,b]))
            print('训练结束！')
            break
```

#### 不使用dataset，使用for分批次取数据
使用epochs

##### 保存模型

```python
x_placeholder = tf.placeholder(dtype=tf.float32, shape=[None,1], name='myInput')
y_placeholder = tf.placeholder(dtype=tf.float32, shape=[None,1])
epochs = 500
batch_size = 64
lr = tf.Variable(0.3, dtype=tf.float32)

w = tf.Variable(np.random.rand(1))
w = tf.cast(w, dtype=tf.float32)
b = tf.Variable(np.random.rand(1))
b = tf.cast(b, dtype=tf.float32)

y_pred = x_placeholder*w + b
tf.identity(y_pred, name='myOutput')

losses = tf.losses.mean_squared_error(y_placeholder, y_pred)
train = tf.train.AdamOptimizer(lr).minimize(losses)


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for i in range(epochs):
        if (i+1)%50==0:
            sess.run(tf.assign(lr, lr/3))
            
        for j in range(int(len(x_data)/batch_size)+1):    # 测试不使用iterator对数据进行批次传入
            x_batch = x_data[j*batch_size:(j+1)*batch_size]
            y_batch = y_data[j*batch_size:(j+1)*batch_size]
            sess.run(train, feed_dict={x_placeholder:x_batch, y_placeholder:y_batch})
            los = sess.run(losses, feed_dict={x_placeholder:x_batch, y_placeholder:y_batch})
            tf.add_to_collection('sum_losses', los)
#         sess.run(train, feed_dict={x_placeholder:x_data, y_placeholder:y_data})
        avg_loss = sess.run(tf.reduce_mean(tf.get_collection('sum_losses')))
        if i%30==0:
            print(i,': ', avg_loss)
        temp = tf.get_collection_ref('sum_losses')
        del temp[:]
    print(sess.run([w,b]))
    tf.saved_model.simple_save(sess, 'test_forBatch_saved_model', inputs={'myInput':x_placeholder}, outputs={'myOutput':y_pred})
```

##### 加载模型

```python
x_test = np.random.rand(10)
x_test = pd.DataFrame(x_test)

with tf.Session(graph=tf.Graph()) as sess:
    tf.saved_model.loader.load(sess, ['serve'], 'test_forBatch_saved_model')
    graph = tf.get_default_graph()
    
    x = sess.graph.get_tensor_by_name('myInput:0')
    y_pred = sess.graph.get_tensor_by_name('myOutput:0')
    print(sess.run(y_pred, feed_dict={x:x_test}))
```
