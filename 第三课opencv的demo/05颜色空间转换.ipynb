#### 颜色空间有四种：RGB、HSV、Lab、Grayscale

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    
def imread(image):
    image = cv2.imread(image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
image = imread('./image/image.jpg')
show(image)
```

### 1、RGB
有红蓝绿三原色组成，范围值为0~255。

```python
(R,G,B) = cv2.split(image)
zeros = np.zeros(image.shape[:2],dtype='uint8')
show(cv2.merge([R, zeros, zeros]))
show(cv2.merge([zeros, G, zeros]))
show(cv2.merge([zeros, zeros, B]))
```

### 2、HSV
HSV是一个比较直观地颜色模型，可以更好的数字化处理颜色，参数为色调(H,Hue), 饱和度(S,Saturation)，明度(V,Value).

```python
image = imread('./image/image.jpg')
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
zeros = np.zeros(image.shape[:2], dtype='uint8')
for name, chan in zip(('H','S','V'), cv2.split(hsv)):
    cv2.imshow(name,chan)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

表示看不出什么效果，只有V的图还能正常看。

### 3、Lab
Lab颜色空间用于计算机色调调整和彩色校正。

```python
image = imread('./image/image.jpg')
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
for name,chan in zip(('L','A','B'), cv2.split(lab)):
    cv2.imshow(name,chan)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

也表示看不出什么效果，只有H的图还能正常看。

### 4、Grayscale
Grayscale是灰度图即黑白图片，通过调节灰度图显示影像。

```python
image = cv2.imread('./image/image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('original', image)
cv2.imshow('gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

这个RGB的红绿蓝三原色图和灰度图可能还有点用。
